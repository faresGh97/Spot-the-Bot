Taking from website https://archive.org/details/booksbylanguage_thai?&sort=-week&page=1

```python
driver.get("https://archive.org/details/booksbylanguage_thai?&sort=-week&page=1")

ScrollNumber = 500
for i in range(1,ScrollNumber):
    driver.execute_script("window.scrollTo(1,50000)")
    time.sleep(5)

file = open('DS.html', 'w',encoding="utf8") 
file.write(driver.page_source)
file.close()

driver.close()
```
