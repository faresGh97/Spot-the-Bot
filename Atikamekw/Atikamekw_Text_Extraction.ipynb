```python
import requests
from bs4 import BeautifulSoup
import random
```

#Processing


### getting all links

```python
processing_flag = True
base_url = "https://atj.wikipedia.org"
query = "/w/index.php?title=Kotakahi:Toutes_les_pages&from=A-ca-oo-mah-ca-ye"
all_links = []
while(processing_flag):
  response = requests.get(
	  url= base_url+query,
  )

  #general defines
  soup = BeautifulSoup(response.content, 'html.parser')
  #finding next Link
  nav_Links = soup.find("div", {"class": "mw-allpages-nav"}).find_all("a")
  next_link = []
  for nav_link in nav_Links:
    if ("minawa masinahikan" in nav_link.text):
      next_link.append(nav_link)

  allLinks = soup.find("div", {"class": "mw-allpages-body"}).find_all("a")
  for link in allLinks:
    # We are only interested in other wiki articles
    all_links.append(link['href'])
  # next link =  next_link[0] if len(next_link) != 0

  if (len(next_link)!= 0):
    query = next_link[0]['href']

  else:
    processing_flag = False

  # #minawa masinahikan (Clare)
  # print (next_link[0])
  # processing_flag = False
print(all_links)
```

```python
# with open("Atikamekw_links.txt", "w") as output:
#     output.write(str(all_links))
with open('Atikamekw_links.txt', 'w') as f:
    for link in all_links:
        f.write(f"{link}\n")
```
