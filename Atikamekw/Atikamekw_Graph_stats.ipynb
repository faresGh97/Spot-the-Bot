```python
bot_paths = []
human_paths = []
counter = 0
with open("./Atikamekw_bot/test_original_corpus.txt") as file_1, open("./Atikamekw_bot/test_bot_corpus.txt") as file_2: 
    human = file_1.readlines()
    bot = file_2.readlines()
    for text1, text2 in zip(human, bot):
        counter += 1
        print (text1[:len(text2)])
        print (text2)
        path_length_text_1 = 0
        path_length_text_2 = 0
        for path in go_through_text(text1 , gabriel):
            path_length_text_1 += len(path)
        for path in go_through_text(text2 , gabriel):
            path_length_text_2 += len(path)

        human_paths.append(path_length_text_1)
        bot_paths.append(path_length_text_2)
        

        
        
```

```python
human_paths = np.array(human_paths)
bot_paths = np.array(bot_paths)
        
X = np.arange(0,counter)
```

```python
print (human_paths)
```

```python
plt.plot(X, human_paths, color='r', label='human')
plt.plot(X, bot_paths, color='g', label='bot')

        # Naming the x-axis, y-axis and the whole graph
plt.xlabel("text number ")
plt.ylabel("Path length")
plt.title("path length comparision for each of human and bot text on SVD Gabreil Graph ")

        # Adding legend, which helps us recognize the curve according to it's color
plt.legend()

        # To load the display window
plt.show()
```
